```python
# Create Filter to identify immunisation percentages
indicatorFilter = ['Immunization, DPT (% of children ages 12-23 months)',]
print(indicatorFilter)
```

```python
# import random to get random samples
import random
# create country filter
countryFilter = random.sample(countries, 2)
```

```python
print(countryFilter)
```

```python
print(indicatorFilter)
```

```python
# create a mesh to apply the country, indicator and years together
filterMesh = (
    (data['CountryName'] == countryFilter[0]) &  # make sure comparison == is in brackets as takes over from & comparitor
    (data['IndicatorName'].isin(indicatorFilter)) &
    (data['Year'].isin(yearsFilter))
)

# note to self make sure that comparison filter is in brackets as overwrist the ampersand
```

```python
# use the mesh to filter data
filteredData1 = data.loc[filterMesh]
```

```python
filteredData1.describe
```

```python
# filter second set of data
# create a mesh to apply the country, indicator and years together
filterMesh = (
    (data['CountryName'] == countryFilter[1]) &  # make sure comparison == is in brackets as takes over from & comparitor
    (data['IndicatorName'].isin(indicatorFilter)) &
    (data['Year'].isin(yearsFilter))
)

filteredData2 = data.loc[filterMesh]
```

```python
filteredData2.describe
```

```python
filteredData2.head()
```

```python
# import plotting to show data of immunisation levels for both countries

# create the stage for plot
fig, ax = plt.subplots(figsize=(20,10))

# set limit for y axis - to base of value data and top of value data
ax.set_ylim(min(0,filteredData1['Value'].min()), 1.2*filteredData1['Value'].max())

# add indicator name title to top of graph
ax.set_title('Indicator Name : ' + indicatorFilter[0])


# plot the data from first country
ax.plot(filteredData1['Year'], filteredData1['Value'], 'r-', label=filteredData1['CountryName'].unique())

# plot the data from second country
ax.plot(filteredData2['Year'], filteredData2['Value'], 'r--', label=filteredData2['CountryName'].unique())



# Add the legend
legend = plt.legend(loc = 'upper center', 
                    shadow=True,
                    prop={'weight':'roman','size':'xx-large'})

# Rectangle around the legend
frame = legend.get_frame()
frame.set_facecolor('.95')
plt.show()
```
